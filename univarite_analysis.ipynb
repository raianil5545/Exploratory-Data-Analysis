```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib as mpl
import numpy as np

mpl.rcParams['figure.dpi']=400
%matplotlib inline
pd.set_option("display.max_columns", None)
```

```python
df=pd.read_csv("final_default_credit.csv")
df.head(5)
```

```python
y = df["default payment next month"].values
```

```python
X = df.drop(columns=["default payment next month"], axis = 1).values
X
```

```python
feature_cols = df.columns.to_list()
feature_cols.remove('default payment next month')
feature_cols
```

```python
#  F-test and Univariate Feature Selection
```

```python
# F test
from sklearn.feature_selection import f_classif

[f_stat, f_p_value] = f_classif(X, y)

f_test_df = pd.DataFrame({"features": feature_cols,
                         "F statistics": f_stat,
                         "p value": f_p_value })
f_test_df.sort_values("p value")
```

```python
# retreving the top 20 percent feature as per F statistics
```

```python
from sklearn.feature_selection import SelectPercentile

selector = SelectPercentile(f_classif, percentile=20)

selector.fit(X, y)
```

```python
best_feature_idx = selector.get_support()
best_feature_idx
```

```python
best_features = [feature_cols[index] for index in range(0, len(feature_cols)) if best_feature_idx[index]] 
```
