```python
import pandas as pd

pd.set_option("display.max_columns", None)
```

```python
default_credit_df = pd.read_excel("default_of_credit_card_clients__courseware_version_1_21_19.xls",
                                )
```

```python
default_credit_df.head(5)
```

```python
columns = list(default_credit_df.columns)
columns
```

```python
# Checking if any columns have a null values
default_credit_df.isnull().sum()
```

```python
default_credit_df.info()
```

```python
# It seems tyhat pay_1 is of an object type but it has got numeric values in it
default_credit_df.PAY_1.unique()
```

```python
# drop the rows where PAY_1 is not available
default_df = default_credit_df[default_credit_df.PAY_1!="Not available"]
default_df["PAY_1"] = pd.to_numeric(default_df.PAY_1, errors="coerce")
default_df.info()
```

```python
# checking the unique values in ID columns
default_df.ID.nunique()
```

```python
# It seems that some IDs are duplicated so need to look further in this matter
print(default_df.ID.value_counts())
# filter out these id with duplicate values
id_counts = default_df.ID.value_counts()
duplicate_id_mask = id_counts != 1
duplicate_id_mask
```

```python
duplicate_ids = id_counts.index[duplicate_id_mask]
len(duplicate_ids)
```

```python
# checking the rows with duplicate values
default_df.loc[default_df["ID"].isin(duplicate_ids), :]
```

```python
# It seems that same ID has one valid data and another with 0 values in all data
df_zero_mask = default_df ==0
```

```python
df_zero_mask
```

```python
# checking the all True in all rows and 1st columns onwards.
feature_zero_mask = df_zero_mask.iloc[:, 1:].all(axis=1)
```

```python
sum(feature_zero_mask)
# it seems that additional rows have all data 0 so we will elimatinate these rows
```

```python
default_clean_df = default_df.loc[~feature_zero_mask, :].copy()
```

```python
default_clean_df.ID.value_counts()
```

```python
default_clean_df.to_csv("clean_default_credit.csv", index=False)
```
